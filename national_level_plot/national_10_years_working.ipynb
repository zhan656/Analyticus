```python
# Dependencies
import json
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Read JSON file
with open('data/plot_national_10_years.json') as plot_national_10_file:
    plot_national_10_dict = json.load(plot_national_10_file)

# Converting JSON data file into DataFrame "df"
df_plot_national_10 = pd.DataFrame(plot_national_10_dict)

# Sort data by week number, start from week 1
df_plot_national_10 = df_plot_national_10.sort_values(by=['year','week'])

# Convert vac_percent to percentage format
# vac_percent = df_plot_national['vac_percent'] * 100
# df_plot_national['vac_percent'] = vac_percent

# Decimal adjustment
df_10 = df_plot_national_10.round({'flu_cases':0, 'flu_percent':2, "vac_percent":2, "vaccinations":0, "week":0})
df_10
```

### 2013

```python
# Filter (Year 2013)
df_10_2013 = df_10.loc[df_10['year']==2013,:]
df_10_2013.head()
```

```python
# set x, y axis
x_axis = df_10_2013["week"]
y_axis_2013_1 = df_10_2013["flu_percent"]
y_axis_2013_2 = df_10_2013["vac_pct_week"]
y_axis_2013_b = df_10_2013["vac_percent"]

# the width of the bars
width = 0.4
```

```python
# Plot bar chart
fig, ax = plt.subplots(figsize=(32, 24), dpi=100, facecolor='w', edgecolor='k')
rects1 = ax.bar(x_axis - width/2, y_axis_2013_1, width, color="Red", label="Percentage of Flu Rate")
rects2 = ax.bar(x_axis + width/2, y_axis_2013_2, width, color="SkyBlue", label="Flu Vaccination Claims Rates")

# Add some text for labels, title and custom x_axis tick labrls
ax.set_xlabel('Week', fontsize = 40)
ax.set_ylabel('Percentage of Flu and Vaccinations', fontsize = 40)
ax.set_title('2013 National Data', fontsize = 50)
ax.set_xticks(x_axis)

ax.legend(loc=7, prop={'size': 30})

def autolabel(rects, xpos="center"):
    # Attach a text label above each bar in *rects*, displaying its height.
    # *xpos* indicates which side to place the text w.r.t. the center of the bar. It can be one of the following {'center', 'right', 'left'}.
    
    # normalize the case of the parameter
    xpos = xpos.lower()
    ha = {'center':'center', 'right':'left', "left":'right'}
    offset = {'center':0.50, 'right':0.57, 'left':0.43}  # x_txt = x + w*off
    
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.01*height, '{}'.format(height), ha=ha[xpos], va='bottom')

autolabel(rects1, "left")
autolabel(rects2, "right")

# Trendline for accumulated vac_percent
# Add second y_Axis to existing chart
ax2 = ax.twinx()
ax2.plot(x_axis, df_10_2013['vac_percent'], color='Green', linewidth = 7.0, label='Accumulated Vaccination Rate')
ax2.set_ylabel('Percentage of accumulated vaccination rate', color='green', fontsize=40)

# Additional legend for trendline
# plt.legend(handles=[ax2], loc=2)
fig.autofmt_xdate()

# Export to PNG file
plt.savefig('National_Data_2013.png')

plt.show()
```

### Number of Flu Cases, Year 2008 - 2017

```python
# Clean data
df_trend = df_10.groupby(['year'])
print(df_trend)
df_year_trendline = df_trend.sum()
df_year_trendline
```

```python
# x, y_axis
x_axis_trendline = ('2008','2009','2010','2011','2012','2013','2014','2015','2016','2017')
y_axis_trendline = df_year_trendline["flu_cases"]
```

```python
# Plot work
fig, ax3 = plt.subplots(figsize=(32, 24), dpi=100, facecolor='w', edgecolor='k')
ax3.plot(x_axis_trendline, y_axis_trendline, color="Cornflowerblue", label="Flu Cases", linewidth = 7.0)
ax3.set_title('Number of Flu Cases, Year 2008 - 2017', fontsize = 40)
ax3.set_xlabel('Year', fontsize = 50)
ax3.set_ylabel('Flu Cases', fontsize = 50)
ax3.spines['bottom'].set_linewidth(5)

ax3.legend(loc=2, prop={'size': 40})

# Tick labels fontsize
plt.tick_params(labelsize=30)

ax3.grid(True)
fig.autofmt_xdate()

# Export to PNG file
plt.savefig('Number of Flu Cases_2008-2017.png')
```
